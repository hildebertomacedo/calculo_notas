```python
#---------------------------------------
#Número de notas
#---------------------------------------

n = int(input("Quantas notas terá a disciplina? "))

#---------------------------------------
#Notas obtidas
#---------------------------------------

#Se tiver 1 nota:
if n == 1:
  #Definindo o valor das notas
  print ("----------------------------------------------------------")
  me_1 = float(input("MEE 1: "))
  print ("----------------------------------------------------------")

  #Definindo o peso de cada nota
  p1 = float(input("Peso (MEE 1): "))

  #Nota * peso
  me1 = me_1 * p1

  #Calculando a média
  mee = (me1)/n
#---------------------------------------

  #Se tiver 2 notas:
elif n == 2:
  #Definindo o valor das notas
  print ("----------------------------------------------------------")
  print("Coloque abaixo suas notas:")
  me_1 = float(input("MEE 1: "))
  me_2 = float(input("MEE 2: "))
  print ("----------------------------------------------------------")

  #Definindo o peso de cada nota
  print("Coloque abaixo o peso das suas notas:")
  p1 = float(input("Peso (MEE 1): "))
  p2 = float(input("Peso (MEE 2): "))

  #Nota * peso
  me1 = me_1 * p1
  me2 = me_2 * p2

  #Calculando a média
  mee = (me1 + me2)/n

#---------------------------------------

  #Se tiver 3 notas:
elif n == 3:
  #Definindo o valor das notas
  print ("----------------------------------------------------------")
  me_1 = float(input("MEE 1: "))
  me_2 = float(input("MEE 2: "))
  me_3 = float(input("MEE 3: "))

  #Definindo o peso de cada nota
  print ("----------------------------------------------------------")
  p1 = float(input("Peso (MEE 1): "))
  p2 = float(input("Peso (MEE 2): "))
  p3 = float(input("Peso (MEE 3): "))


  #Nota * peso
  me1 = me_1 * p1
  me2 = me_2 * p2
  me3 = me_3 * p3

  #Calculando a média
  mee = (me1 + me2 + me3)/n

else:
    print("Digite uma quantidade válida")

#---------------------------------------
#Condições (Aprovado / Reprovado)
#---------------------------------------

if (0 < n < 10):
 #Aprovado
    if (mee) >= 8:
        print ("----------------------------------------------------------")
        print("Parabéns! Você passou direto.")

#Aprovado com direito a PF
    elif 7.5 <= (mee) < 8:
        mf = mee
        print ("----------------------------------------------------------")
        print ("Você passou com: ", '{0:.2f}'.format(mf), "Recomenda-se fazer a PF.")

#Reprovado por nota, precisa fazer a PF
    elif (mee) < 7.5:
        pf = 0
        mf = ((mee*2)+pf)/n
        print ("----------------------------------------------------------")
        print ("Você precisa fazer a PF. Sua nota foi de: ", '{0:.2f}'.format(mf))

  #Calculando o necessário para ser aprovado:
        p = 15-(mee*2)
        print ("----------------------------------------------------------")
        print ("Você precisa tirar acima de", '{0:.2f}'.format(p), "ponto(s) na PF para ser aprovado.")

  #Substituindo pela nota da pf
        print ("----------------------------------------------------------")
        nova_pf = input("Deseja inserir o valor real da PF? s = sim / n = não ")
        if nova_pf == "s":
            print ("----------------------------------------------------------")
            pf = float(input("Insira o valor real da PF: "))
            mf = ((mee*2)+pf)/3
            print ("----------------------------------------------------------")
            print ("Nova média final: ", '{0:.2f}'.format(mf))
            if mf < 5:
                print ("----------------------------------------------------------")
                print ("Sinto muito! Você está reprovado.")
            else:
                print ("----------------------------------------------------------")
                print ("Parabéns! Você está aprovado.")
        if nova_pf == "n":
            print ("----------------------------------------------------------")
            print ("Código finalizado")
```
